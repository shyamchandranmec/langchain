```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.prompts import (
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate, 
    PromptTemplate,
    ChatPromptTemplate
    )
from langchain_ollama import ChatOllama
```

```python
model = 'qwen3'
base_url = "http://localhost:11434"
llm = ChatOllama(
    model = model,
    base_url = base_url
)
```

```python
system_message_prompt_template = SystemMessagePromptTemplate.from_template("You are {character} in One Piece Anime. Respond in their personality always")
human_message_prompt_template =HumanMessagePromptTemplate.from_template("Tell me about {topic} in {points} short points")
chat_prompt_template = ChatPromptTemplate.from_messages([system_message_prompt_template,human_message_prompt_template])  

#without chains
chat_prompt_value = chat_prompt_template.invoke({"topic": "Sun", "character": "Robin", "points": 3})
response = llm.invoke(chat_prompt_value)
response.content
```

```python
#with chains
chain = chat_prompt_template | llm
response = chain.invoke({"topic": "Earth", "character": "Shanks", "points": 3})
print(response.content)
```

```python
#with output parser
from langchain_core.output_parsers import StrOutputParser

chain = chat_prompt_template | llm | StrOutputParser()
response = chain.invoke({"topic": "Earth", "character": "GolD Roger", "points": 3})
print(response)
```

```python
analysis_chat_prompt = ChatPromptTemplate.from_template(
    """
    You ar Big Mom from One piece. Analyse the following text and 
    tell me how difficult it is to understand in one sentence.
    {text}
    """
)

fact_check_chain = analysis_chat_prompt | llm | StrOutputParser()
output = fact_check_chain.invoke({"text": chain})
output
```

```python
composed_chain = {"text": chain} | fact_check_chain
output = composed_chain.invoke({"topic": "Earth", "character": "Black Beard", "points": 3})
output
```
