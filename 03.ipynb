```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.prompts import (
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate, 
    PromptTemplate,
    ChatPromptTemplate
    )
from langchain_ollama import ChatOllama
```

```python
model = 'qwen3'
base_url = "http://localhost:11434"
llm = ChatOllama(
    model = model,
    base_url = base_url
)
```

```python
system_message_prompt_template = "You are {character} in One Piece Anime. Respond in their personality always"
human_message_prompt_template = "Tell me about {topic} in {points} points"
chat_prompt_template = ChatPromptTemplate.from_messages([system_message_prompt_template, human_message_prompt_template])

#without chains
chat_prompt_value = chat_prompt_template.invoke({"topic": "Sun", "character": "Chopper", "points": 3})
response = llm.invoke(chat_prompt_value)
response.content
```

```python
#with chains
chain = chat_prompt_template | llm
response = chain.invoke({"topic": "Earth", "character": "Shanks", "points": 3})
print(response.content)
```

```python
#with output parser
from langchain_core.output_parsers import StrOutputParser

chain = chat_prompt_template | llm | StrOutputParser()
response = chain.invoke({"topic": "Earth", "character": "GolD Roger", "points": 3})
print(response)
```
