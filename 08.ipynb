```python
from dotenv import load_dotenv
from langchain_ollama import ChatOllama
from typing import Optional
from pydantic import BaseModel, Field
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import (ChatPromptTemplate,
                                    PromptTemplate,
                                    SystemMessagePromptTemplate,
                                    HumanMessagePromptTemplate)
```

```python
load_dotenv()

llm =ChatOllama(
    base_url="http://localhost:11434",
    model = "qwen3"
)
```

```python
template = ChatPromptTemplate.from_template("{prompt}")
chain = template | llm | StrOutputParser()
about = "My name is Shyam and I am a software developer. I have experience in Python, JavaScript, and cloud technologies. I enjoy learning new programming languages and frameworks."
response = chain.invoke({"prompt":about})
print(response)
```

```python
prompt = "What is my name?"
response = chain.invoke({"prompt":prompt})
print(response)
```

```python
from langchain_core.messages import (HumanMessage, SystemMessage, AIMessage)
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import SQLChatMessageHistory
```

```python
def get_session_history(session_id):
    return SQLChatMessageHistory(session_id, "sqlite:///chat_history.db")
```

```python
runnable_with_history = RunnableWithMessageHistory(chain, get_session_history)
user_id = "shyam_id"
history = get_session_history(user_id)
```

```python
runnable_with_history.invoke([HumanMessage(content=prompt)], config={
    "configurable": {'session_id': user_id}
})
```
