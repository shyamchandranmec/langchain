```python
from dotenv import load_dotenv
from langchain_ollama import ChatOllama
from langchain_core.messages import (HumanMessage, SystemMessage)
from langchain_core.prompts import (
    HumanMessagePromptTemplate,
    SystemMessagePromptTemplate,
    ChatPromptTemplate,
    PromptTemplate
)
from langchain_core.output_parsers import StrOutputParser
```

```python
llm = ChatOllama(
    base_url = "http://localhost:11434",
    model = "qwen3"
)

smpt = SystemMessagePromptTemplate.from_template("You are a {school} teacher")
hmpt = HumanMessagePromptTemplate.from_template("Tell me about {topic} in {points} points")
messages = [smpt, hmpt]
cpt = ChatPromptTemplate.from_messages(messages)
fact_chain = cpt | llm | StrOutputParser()
output = fact_chain.invoke({"topic": "AI", "points": 3, "school": "phd"})
print(output)
```

```python
hmpt = HumanMessagePromptTemplate.from_template("Write a poem on  {topic} in {sentences} lines")
messages = [smpt, hmpt]
cpt = ChatPromptTemplate.from_messages(messages)
poem_chain = cpt | llm | StrOutputParser()
output = poem_chain.invoke({"topic": "Earth", "sentences": 3, "school": "phd"})
print(output)
```

```python
from langchain_core.runnables import RunnableParallel
chain = RunnableParallel(fact = fact_chain, poem = poem_chain)
output = chain.invoke({"topic": "AI", "points": 3, "sentences": 3, "school": "phd"})
print(output['fact'])
print(output['poem'])
    
```
