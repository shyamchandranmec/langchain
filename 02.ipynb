```python
from dotenv import load_dotenv
from langchain_ollama import ChatOllama
load_dotenv()
base_url = "http://localhost:11434"
model = "qwen3"
```

```python
from langchain_core.messages import SystemMessage, HumanMessage
llm = ChatOllama(
    base_url =  base_url,
    model = model
)
human_message = HumanMessage("Tell me about the earth in 3 points")
#system = SystemMessage("You are an elementary teacher. You answer in short sentences")

system_message = SystemMessage("You are Luffy from One piece anime. Act as an assistant. Respond with his precise intellect, tone, logic and humour. Maintain his speech patterns, catchphrases and confidence in his intellect at all times")

messages = [system_message, human_message]
response = llm.invoke(messages)
print (response.content)

```

```python
from langchain_core.prompts import(
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,
    ChatPromptTemplate,
    PromptTemplate
)
```

```python

system_prompt_template = SystemMessagePromptTemplate.from_template("You are {character} from One piece anime. Act as an assistant. Respond with his precise intellect, tone, logic and humour. Maintain his speech patterns, catchphrases and confidence in his intellect at all times")
human_prompt_template = HumanMessagePromptTemplate.from_template("Tell me about the {topic} in {points} points")
system_prompt_template
```

```python
#method 1
system_message = system_prompt_template.format(character = "Zoro")
human_message  = human_prompt_template.format(topic = "Earth", points = 2)
system_message
```

```python
messages = [system_message, human_message]

```

```python
response = llm.invoke(messages)
response.content
```

```python
## This is wrong way of doing the same thing. There is no point in putting already formated messages in chat prompt template
## Not removing it so that I won't forget that this is wrong way of doing the same thing
## The typical use case for ChatPromptTemplate is when you build it from template   objects (with {variables}), so it can be reused with different inputs:
                                                                         
chat_prompt_template = ChatPromptTemplate.from_messages(messages)
chat_prompt_template
response = llm.invoke(chat_prompt_template.messages)
response.content
```

```python
#method 3
prompt_templates = [system_prompt_template, human_prompt_template]
chat_prompt_template = ChatPromptTemplate.from_messages(prompt_templates)
chat_prompt_value  = chat_prompt_template.invoke({'character':"Nami", 'topic':"Weather", "points":2})
response = llm.invoke(chat_prompt_value)
print(response.content)
```

```python
#method 4
prompt_templates = [system_prompt_template, human_prompt_template]
chat_prompt_template = ChatPromptTemplate.from_messages(prompt_templates)
chat_prompt_value  = chat_prompt_template.format_messages(character="Nami", topic="Weather", points=2)
response = llm.invoke(chat_prompt_value)
print(response.content)
```
